```python
import pandas as pd 
import numpy as np
data = pd.read_csv('12.27.2020_WaterData.csv',sep='\t')
reference = pd.read_excel('C:/Users/Scott/Desktop/Projects/River_Data/RiverReferenceTable.xlsx')
df=data.set_index('Name').join(reference.set_index('USGS Name'))
df.reset_index(inplace=True)
df['Value'] = df.Value.apply(lambda x:int(x.replace(',','')))
df['Date'] = pd.to_datetime(df.Date)
```

```python
Discharge = df[['Value','Date','Name',
                'OptimalDischargeMin_Editable','OptimalDischargeMax_Editable',
               'FishType']].loc[df.VariableDescription=='Discharge']
Discharge.head()
```

```python
DischargePiv = pd.pivot_table(data=Discharge,index='Date',columns='Name',values='Value',
              aggfunc=np.mean).sort_values(by='Date')
DischargePiv.head()
```
